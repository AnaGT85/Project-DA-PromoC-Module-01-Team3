## Lectura documento txt. 
Nos falta realizar una funcion para poder hacer todos los replace juntos

```python
#Sacamos el archivo

with open('data_txt_reorder.txt','r') as file:
    file_source = file.readlines() # Es una lista que cada línea del fichero txt como un string de todos los datos.
```

## Realizaremos los cambios pedidos:
    Contiene espacios extra en algunos strings, que deben ser eliminados

    Contiene elementos como '\n' que deben ser eliminados

    Contiene elementos como '<' que deben ser cambiados por 'under'
    
    Contiene elementos como null, que deben ser cambiados por 'NULL'

```python
#Realizamos una copia para poder trabajar con el archivo 

copia1 = file_source.copy()
```

```python
#Empezamos quitando los \n del archivo

lista2 = []
for i in copia1[0:]:
    x = i.replace("\n", "")
    lista2.append(x)


```

```python
#Seguimos cambiando null por NULL

lista3 = []
for i in lista2[0:]:
    x = i.replace("null", "NULL")
    lista3.append(x)
```

```python
#Reemplazamos el < por under

lista4 = []
for i in lista3[0:]:
    x = i.replace("<", "under")
    lista4.append(x)
```

```python
#hemos creado una funcion para limpiar el documento de espacios vacios en los string
lista_vacia= [] #Hacemos una lista vacia para poder apendear los resultados de la limpieza

def limpiar_espacios(str):
    brocoli=str.split(",")
    for i in brocoli :
        final= i.strip()
        lista_vacia.append(final)
    return lista_vacia
    

    
```

```python
for i in lista4:
    x = limpiar_espacios(i)
    lista_vacia.append(x)

lista_vacia #comprobamos que se han realizado todas las mejoras
```
